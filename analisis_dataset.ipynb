```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os


# Load the dataset as one DataFrame for all CSV files in the 'data' folder
def load_data(data_folder):
    all_files = [f for f in os.listdir(data_folder) if f.endswith('.csv')]
    data_frames = []
    for file in all_files:
        print(f"Loading file: {file}")
        file_path = os.path.join(data_folder, file)
        df = pd.read_csv(file_path, sep=';', encoding='latin')
        data_frames.append(df)
    
    return pd.concat(data_frames, ignore_index=True)

df = load_data('data')
```

```python
# save dataset only with:
# FECHA_PROCESO as datetime
# ACUERDO_MARCO as string
# ENTIDAD as string
# PROVEEDOR as string
# TIPO_PROCEDIMIENTO as string
# SUB_TOTAL as float
# TOTAL as float
# IGV as float

# yyyy-mm-dd hh:mm:ss
df['FECHA_PROCESO'] = pd.to_datetime(df['FECHA_PROCESO'], errors='coerce', format='%Y-%m-%d %H:%M:%S')
df['ACUERDO_MARCO'] = df['ACUERDO_MARCO'].astype(str)
df['ENTIDAD'] = df['ENTIDAD'].astype(str)
df['PROVEEDOR'] = df['PROVEEDOR'].astype(str)
df['TIPO_PROCEDIMIENTO'] = df['TIPO_PROCEDIMIENTO'].astype(str)
df['SUB_TOTAL'] = pd.to_numeric(df['SUB_TOTAL'], errors='coerce')
df['TOTAL'] = pd.to_numeric(df['TOTAL'], errors='coerce')
df['IGV'] = pd.to_numeric(df['IGV'], errors='coerce')
# Save the cleaned dataset
df_cleaned = df[['FECHA_PROCESO', 'ACUERDO_MARCO', 'ENTIDAD', 'PROVEEDOR', 'TIPO_PROCEDIMIENTO', 'SUB_TOTAL', 'TOTAL', 'IGV']]

# Save the cleaned DataFrame to a new CSV file
df_cleaned.to_csv('ReportePCBienes202201_cleaned.csv', index=False, encoding='utf-8')
```
